```python
# problem 1
#-------------------------------------------------------------------------------------------------------------------------------

import matplotlib.pyplot as plt
import numpy as np
import re
from collections import defaultdict



freqs = {}

with open('test.txt') as f:
    for line in f:
        line= "".join(c.lower() for c in line if c.isalpha())
        for char in line:
            if char in freqs:
                    freqs[char] += 1
            else:
                    freqs[char] = 1

print(freqs)

sorted_keys = freqs.items()
new_values = sorted(sorted_keys)
sorted_freqs = {}

for i in range(len(new_values)):
    sorted_freqs[new_values[i][0]] = new_values[i][1]

print(sorted_freqs)



summ= sum(freqs.values())


print('summ of val', summ)




x = sorted_freqs
keys = x.keys()
vals = x.values()

plt.bar(keys, list(vals), label=' Un norm fREQ')

plt.ylim(0,)
plt.ylabel ('key')
plt.xlabel ('Friq')
plt.xticks(list(keys))
plt.legend (bbox_to_anchor=(1, 1), loc="upper right", borderaxespad=0.)

plt.show()

print('--------------------------------------------------------------------------------------------------------\n')

```
